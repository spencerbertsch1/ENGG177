```python
import random
import matplotlib.pyplot as plt
plt.style.use("fivethirtyeight")
from matplotlib.pyplot import figure
%config InlineBackend.figure_formats = ['svg']
%matplotlib inline
```

```python
def up_or_down():
    """
    utility function to get whether or not the security will go up or down 
    """
    val = random.uniform(0, 1)
    if val >= 0.45:
        return True
    else:
        return False
```

```python
simulations: int = 30
figure(figsize=(16, 6), dpi=80)
for i in range(simulations):
    
    iterations = 300
    capital: int = 50
    curr_capital = []
    for i in range(iterations):

        if up_or_down() is True: 
            # our capital increases by 5%
            capital = capital + capital*0.05
        else:
            # our capital decreases by 10%
            capital = capital - capital*0.10

        curr_capital.append(capital)
        
    x = list(range(iterations))
    y = curr_capital
    
    plt.plot (x, y, linewidth=1)
    plt.title('Projected IBM Stock Value Over Time', fontsize=18)
    plt.ylabel('IBM Stock Value', fontsize=14)
    plt.xlabel('Days (N=300)', fontsize=14)
    
plt.show()
```
